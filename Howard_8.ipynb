```python
import pandas as pd
import seaborn as sns
```

```python
avocados = pd.read_csv('avocado.csv')
avocados.head()                  
```

```python
avocados.info()
```

```python
avocados['Date'] = pd.to_datetime(avocados['Date'])
avocados.info()
```

```python
avocados.query('year == 2015 and type == "conventional" and region == "Albany"')
avocados_melt = avocados[['Date', 'Total Bags', 'Small Bags']]
avocados_melt.head()
```

```python
avocados_plot = pd.melt(avocados_melt, id_vars = 'Date', value_vars = ['Total Bags', 'Small Bags'], var_name = "Bags", value_name = "Count")
display(avocados_plot)
```

```python
sns.relplot(data = avocados_plot, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
```

```python
avocados['Sales Volume'] = pd.qcut(avocados['Total Volume'], q=4, labels = ['Poor', 'Modest', 'Good', 'Excellent'])
avocados.head()
```

```python
sns.countplot(data = avocados, x = 'year', hue = 'Sales Volume')
```

```python
avocados_plot = pd.melt(frame = avocados, id_vars = 'year', value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'], var_name = "Bags", value_name = "Count")
display(avocados_plot)
```

```python
sns.relplot(data = avocados_plot, kind = 'line', x = 'year', y = 'Count', hue = 'Bags')
```
